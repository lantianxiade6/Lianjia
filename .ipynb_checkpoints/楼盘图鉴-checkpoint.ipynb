```python
label=[str(i) for i in data2['district'].unique()]#取出所有district名
label
def getval(stri='毛坯',label=label):
    mp=by_dec.iloc[by_dec.index.get_level_values(2).str.contains(stri)]#取出毛坯数据
    mp2=mp.reset_index(level=[0,2],drop=True)#只保留district
    value=[]
    for val in mp2[label]:#按label顺序取出average_price
        if val==val:#如果该区对应的average_price非nan
            value.append(int(val))#存入average_price
        else:
            value.append(0)#否则存入0
    return value
```

```python
from pyecharts.charts import WordCloud
from collections import Counter

def myTags(city=['广州市','佛山市']):
    alltag = []
    for st in data2[(data2['city_name'].isin(city))].dropna(subset=['tags'])['tags']:#取出所有非空的house_tag
        alltag.extend(st)#存入一个统一的list

    freq=[(name,value) for name,value in Counter(alltag).items()]#计算每个元素的频数
    #print(freq)
    wordcloud=WordCloud()
    wordcloud.add("",freq)#按频数出词云
    return wordcloud.render_notebook()

myTags()#标签结果，不分城市
```

```python
with open('stoplist.txt','r') as f:#读取停词
    stop=f.readlines()
stop=[i.strip('\n') for i in stop]
```

```python
import jieba
import re
def myDesc(city=['广州市','佛山市']):
    alltag = []
    for st in data2[(data2['city_name'].isin(city))].dropna(subset=['project_desc'])['project_desc']:#取出所有非空的project_desc
        alltag.extend(jieba.cut(st))#直接切词存入一个统一的list
    alltag=[word for word in alltag if word not in stop]#去掉停词
    alltag=[re.sub("[A-Za-z0-9]", "", word) for word in alltag]#去掉字母数字
    alltag=[word for word in alltag if len(word)>1]#自保留两个字以上的词
    freq=[(name,value) for name,value in Counter(alltag).items()]#计算每个元素的频数
    #print(freq)
    wordcloud=WordCloud()
    wordcloud.add("",freq)#按频数出词云
    return wordcloud.render_notebook()

myDesc()#描述结果，不分城市
```

平均建筑面积

```python
#平均建筑面积
data2['avg_frame_area']=(data2['min_frame_area']+data['max_frame_area'])/2
data2.head()
```

```python
fig = plt.figure(dpi=300)
city_avg_frame=data2[data2['avg_frame_area']>0].groupby(['city_name','district'])['avg_frame_area'].mean()#分城市均价
city_avg_frame.plot.barh()
```

```python
def clean_area_range(area_range):
    result=re.sub("m|㎡","",area_range)#删掉m或m方
    result=result.split("-")#按“-”分割
    result=[float(i) for i in result]#转数值
    result=np.mean(result)#算均值
    return result
```

```python
#拉平converged_rooms，算每个房的平均面积
city=[]
district=[]
bedroom_count=[]
area_range=[]
for row in data2.index:#遍历每一行
    for item in data2['converged_rooms'][row]:#遍历converged_rooms中的每一个
        if 'bedroom_count' in item and 'area_range' in item:
            city.append(data2['city_name'][row])#存入城市
            district.append(data2['district'][row])#存入区
            bedroom_count.append(item['bedroom_count'])#存入房数
            area_range.append(clean_area_range(item['area_range']))#存入平均面积
```

```python
frame=pd.DataFrame({'city':city,'district':district,'bedroom_count':bedroom_count,'area_range':area_range})
```

```python
fig = plt.figure(dpi=300)
city_avg__frame=frame[frame['area_range']>0].groupby(['city','district'])['area_range'].mean()#分城市分区平均面积
city_avg__frame.plot.barh()
```

```python
fig = plt.figure(dpi=300)
#分城市分区分房型平均面积
city_avg_bedroom_frame=frame[(frame['area_range']>0) & (frame['bedroom_count']!='0')].groupby(['city','bedroom_count'])['area_range'].mean()
city_avg_bedroom_frame.plot.barh()
```

距离

```python
#计算经纬度换算成距离（貌似还不太准确）
#参考网址：https://www.cnblogs.com/lgh344902118/p/7490795.html
import math
EARTH_REDIUS = 6378.137
def rad(d):
    return d * math.pi / 180.0

def getDistance(lat1, lng1, lat2, lng2):
    radLat1 = rad(lat1)
    radLat2 = rad(lat2)
    a = radLat1 - radLat2
    b = rad(lng1) - rad(lng2)
    s = 2 * math.asin(math.sqrt(math.pow(math.sin(a/2), 2) + math.cos(radLat1) * math.cos(radLat2) * math.pow(math.sin(b/2), 2)))
    s = s * EARTH_REDIUS
    return s
```

```python
distance=[]
for row in data2.index:
    distance.append(getDistance(23.125475,113.255686,data2['latitude'].loc[row],data2['longitude'].loc[row]))#计算和西门口的距离（km）
    
data2['distance']=distance
```
